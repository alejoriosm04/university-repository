# **Criptografía con matrices**

La criptografía es la técnica de escribir procedimientos o claves secretas,  de tal forma que los escritos solo sean inteligible para quien sepa descifrarlo.

Conceptos básicos:

1. **Cifrado**: El procedimiento que generará un mensaje ininteligible para el receptor.
2. **Texto plan**o: El mensaje o información que se va a codificar.
3. **Texto cifrad**o: El mensaje o información que se obtiene después que se ha utilizado el cifrado.


El sistema de cifrado "**Hill**" utiliza una matriz como un sistema de cifrado para codificar mensajes y es muy difícil de romper cuando se utiliza una matriz de gran tamaño.

## **Método Hill**

Este método funciona utilizando la primera matriz (matriz de codificación) y su inversa llamada matriz de decodificación.

Supongamos que tenemos una matriz invertible $A$, la cual será la matriz de codificación.  Transformamos el texto a codificar en una serie de números dando a cada letra  un único valor numérico. A continuación, formaremos una matriz mediante la agrupación de los números en columnas conforme al orden de la matriz $A$  (la cantidad de elementos debe ser igual al orden de la matriz).

Llamemos $B$ a la matriz plana (matriz que contiene el mensaje que se desea codificar). Multipliquemos $AB = C,$ donde $C$ será la matriz codificada.
Para decodificar el  mensaje se debe realizar el producto  $A^{-1}C,$ donde $A^{-1}$ es la matriz inversa de $A$.

El texto plano original se puede hallar nuevamente tomando la matriz resultante y uniendo sus columnas de manera que formen una secuencia y posteriormente convertir los números en caracteres respectivos.  

**Codificación:**

Para codificar un mensaje, podemos enumerar todas las letras del alfabeto como se muestra en la siguiente tabla (módulo 27):

\

\

Aquí se asume que a los espacios entre cada palabra se codifican con el número 28. En estos términos:

1. Escogemos la llave que usaremos para codificar el mensaje. En este ejemplo usaremos la llave: 2, 2, 4, -2, 4, -2, -2, -4, -4, y la dimensión de la matriz será $3 \times 3$, es decir,



\

\begin{equation*}
 A =
\begin{pmatrix}
2 & 2 & 4\\
-2 & 4 & -2\\
-2 & -4 & -4
\end{pmatrix}.
\end{equation*}

\


El mensaje a codificar es : **QUIMICA** que corresponde a los números 18, 22, 9, 13, 9, 3, 1. Lo colocamos en la matriz $B$ y en los huecos restantes escribimos $0,$ es decir,

\

\begin{equation*}
 B =
\begin{pmatrix}
18 & 22 & 9\\
13 & 9 & 3\\
1 & 0 & 0
\end{pmatrix}.
\end{equation*}



\


Ahora, realizamos el producto $AB = C$:


\




\begin{equation*}
AB =
\begin{pmatrix}
2 & 2 & 4\\
-2 & 4 & -2\\
-2 & -4 & -4
\end{pmatrix}
\begin{pmatrix}
18 & 22 & 9\\
13 & 9 & 3\\
1 & 0 & 0
\end{pmatrix}
=
\begin{pmatrix}
66 & 62 & 24\\

\end{pmatrix}
= C.
\end{equation*}


\


Por lo tanto, el mensaje encriptado es: 66, 62, 24, 14, -8, -6, -92, -80, -30.

**Decodificación:**

Para decodificar el mensaje es necesario conocer la matriz $A$, y por ende su inversa, que en este caso es dada por:

\

\begin{equation*}
 A^{-1} =
\begin{pmatrix}
-3 & -1 & -5/2\\
-1/2 & 0 & -1/2\\
2 & 1/2 & 3/2
\end{pmatrix}.
\end{equation*}

\


Finalmente, realizamos el producto $A^{-1}C$:

\



\begin{equation*}
A^{-1}C =
\begin{pmatrix}
-3 & -1 & -5/2\\
-1/2 & 0 & -1/2\\
2 & 1/2 & 3/2
\end{pmatrix}
\begin{pmatrix}
66 & 62 & 24\\
14 & -8 & -6\\
-92 & -80 & -30
\end{pmatrix}
=
\begin{pmatrix}
18 & 22 & 9\\
13 & 9 & 3\\
1 & 0 & 0
\end{pmatrix}
= B.
\end{equation*}


\


La escribimos en forma de texto: 18, 22, 9, 13, 9, 3, 1, 0, 0. Y por último usando la tabla anterior obtenemos el mensaje que es QUIMICA.


**Ejemplo 1:** María le ha enviado el siguiente mensaje codificado a Juan:

14, 15, 18, -10, 21, 106, 69, 2, 122, 83, 128, 101, 2, 100, 109, -86, -74, 18, -72, -86, 66, 27, -8, 66, 31.

Usando la siguiente información decodifique el mensaje:

1. Llave de codificación: 0, 0, 1, -1, 0, 0, 4, -1, 1, 2, -2, 4, 1, 1, 2, 2, -4, 0, 0, -2, 0, 2, -1, 1, 2.
2. Dimensión de la matriz de codificación $5 \times 5$.

**Solución:** De acuerdo a la información suministrada, la matriz asociada a la llave es la matriz $5\times 5$:

\


\begin{equation*}
 A =
\begin{pmatrix}
0 & 0 & 1 & -1& 0\\
0 & 4& -1 & 1 & 2\\
-2 & 4 & 1 & 1 & 2\\
2 & -4 & 0 & 0 & -2\\
0 & 2 & -1 & 1 & 2
\end{pmatrix}.
\end{equation*}



\

Por otro lado, la matriz correspondiente al mensaje codificado es:

\begin{equation*}
 C =
\begin{pmatrix}
14 & 15 & 18 & -10 & 21\\
106 & 69 & 2 & 122 & 83\\
128 & 101 & 2 & 100 & 109\\
-86 & -74 & 18 & -72 & -86\\
66 & 27 & -8 & 66 & 31
\end{pmatrix}.
\end{equation*}

```python
# Construcción de las matrices A y C.

# Importamos la biblioteca numpy con el alias np.
import numpy as np

# Definimos las matrices A y C

A = np.array([[0, 0, 1, -1, 0],[0, 4, -1, 1, 2],[-2, 4, 1, 1, 2],[2, -4, 0, 0, -2],[0, 2, -1, 1, 2]])

C = np.array([[14, 15, 18, -10, 21],[106, 69, 2, 122, 83],[128, 101, 2, 100, 109],[-86, -74, 18, -72, -86],[66, 27, -8, 66, 31]])

```

Para descifrar el mensaje se requiere multiplicar la matriz $C$ por la inversa de $A$.

```python
# Calculo de la inversa de A

A_inv = np.linalg.inv(A)

# Multiplicación B = (A^-1)*C

B = A_inv @ C

print(B)
```

Por lo tanto, el mensaje corresponde a la secuencia:

17, 5, 19, 20, 9, 20, 21, 5, 28, 26, 28, 21, 19, 9, 22, 14, 6, 1, 19, 1, 20, 0, 0, 0, 0.

En estos términos, el mensaje que María le envío a Juan es: **PERSISTE Y VENCERAS**



## **Sistemas de ecuaciones lineales**

**Ejemplo 3:**
Una compañía minera extrae mineral de dos minas, el cual contiene para la mina $I$ el $1\%$ de níquel y $2\%$ de cobre, para la mina $II$ el $2\%$ de níquel y $5\%$ de cobre. ¿Qué cantidad de mineral se deberá extraer de cada mina para obtener $4$ toneladas de níquel y $9$ toneladas de cobre?

**Solución:** Sean
$$x =   ``\mbox{el número de toneladas que se extrae de la mina I}", $$
$$ y =  ``\mbox{el número de toneladas que se extrae de la mina II}". $$

Así, para saber cuántas toneladas hay que extraer de cada mina debemos resolver el sistema ecuaciones lineales:

\begin{eqnarray}
0.01x + 0.02y = 4 \\
0.02x + 0.05y = 9.
\end{eqnarray}

```python
# Solución del Ejemplo 2.

# Definimos la matriz A asociada al sistema y el vector b

A = np.array([[0.01, 0.02], [0.02, 0.05]])

b = np.array([[4],[9]])

# Resolviendo sistema de ecuaciones
x = np.linalg.solve(A, b)

print(x)
```

Por lo tanto, se deben de extraer $200$ toneladas de la mina $I$ y $100$ toneladas de la mina $II.$


**Ejemplo 3:** Juan invierte $\$ 9,700$ en dos cuentas bancarias. La primera cuenta pagó $7 \%$, la segunda cuenta pagó $11 \%$ en intereses. Al finalizar el primer año habia ganado $\$ 863$ en intereses. ¿Cuánto habia en cada cuenta?


**Solución:** Sean
$$x =   ``\mbox{la cantidad de inversión en la primera cuenta}", $$
$$ y =  ``\mbox{la cantidad de inversión en la segunda cuenta}". $$

De esta manera, podemos resumir la información en una tabla como se muestra a continuación:

\

\begin{array}{|c|c|} \hline
 & \mbox{Dinero invertido} & \mbox{Tasa}  & \mbox{Total de intereses} \\ \hline
\mbox{Cuenta} \ 1 & x  & 0.07 & 0.07x \\
\mbox{Cuenta} \ 2 & y & 0.11 & 0.11y \\
\mbox{Total} & \$ 9,700 &  & \$ 863 \\ \hline
\end{array}

\\

De la tabla anterior, podemos observar que para dar respuesta a dicho problema es necesario resolver el sistema de ecuaciones:

\

\begin{eqnarray}
x + y = 9700 \\
0.07x + 0.11y = 863.
\end{eqnarray}


```python
# Solución del Ejemplo 3.

# Definimos la matriz A asociada al sistema y el vector b

A = np.array([[1, 1], [0.07, 0.11]])

b = np.array([[9700],[863]])

# Resolviendo sistema de ecuaciones
x = np.linalg.solve(A, b)

print(x)
```

En consecuencia, la cantidad de dinero invertida en la cuenta 1 y 2 fueron de \$5,100 y   \$ 4,600 respectivamente.

**Ejemplo 4:** Una empresa de transporte gestiona una flota de $60$ camiones de tres modelos diferentes. Los mayores transportan una media diaria de $ 1,500 \ kg$ y recorren diariamente una media de $400 \ km$. Los medianos transportan diariamente una media de  $5,000 \ kg$ y recorren $300 \ km$. Los pequeños transportan diariamente  $ 5,000 \ kg$ y recorren $100 \ km$ de media. Diariamente los camiones de la empresa transportan un total  de $475,000 \ kg$ y recorren $12,500 \ km$ entre todos. ¿Cuántos camiones gestiona la empresa de cada modelo?  

**Solución:** Definamos  
$$x =   ``\mbox{cantidad de camiones grandes}", $$
$$ y =  ``\mbox{cantidad de camiones medianos}", $$
$$ z =  ``\mbox{cantidad de camiones pequeños}". $$

La información suministrada en el problema se puede condensar en la siguiente tabla:

\

\begin{array}{|c|c|} \hline
\mbox{Tipo de camión} & \mbox{Número de camiones} & \mbox{Kilogramos diarios}  & \mbox{Kilómetros diarios} \\ \hline
\mbox{Grandes} & x  & 15,000x & 400x \\
\mbox{Medianos}  & y & 10,000y & 300y \\
\mbox{Pequeños} & z & 5,000z & 100z \\
\mbox{Total} & 60 & 475,000 & 12,500 \\
 \hline
\end{array}

\\

De la tabla anterior, podemos observar que para dar respuesta a dicho problema es necesario resolver el sistema de ecuaciones:

\

\begin{eqnarray}
x + y + z = 60 \\
15,000x + 10,000y + 5,000z = 475,000\\
400x + 300y + 100z = 12,500
\end{eqnarray}

```python
# Solución del Ejemplo 4.

# Definimos la matriz A asociada al sistema y el vector b

A = np.array([[1, 1, 1], [15000, 10000, 5000],[400, 300, 100 ]])

b = np.array([[60],[475000],[12500]])

# Resolviendo sistema de ecuaciones
x = np.linalg.solve(A, b)

print(x)
```

Por lo tanto, el número de camiones grandes es de 5, el de camiones medianos es de 25 y el total de camiones pequeños es de 30.
