### **Ejercicio #1: (0.8)**
Considere la matriz

\begin{equation*}
 A =
\begin{pmatrix}
 -k & k-1 & k+1 & k\\
1 & 2 & 3 & 4\\
2-k & k+3 & k+7 & 2 \\
-2 & 3 & 4 & 6
\end{pmatrix}.
\end{equation*} \


a). **(0.4)**  Halle los valores de $k$ para los cuales la matriz $A$ tiene inversa.

b). **(0.4)**  Halle los valores de $k$ para los cuales $\det(A) = 1-7k$.


```python
# Solución Ejercicio 1. (No olvide imprimir las respuestas de sus cálculos)

# (a):
import sympy as sp

k = sp.Symbol('k')

A = sp.Matrix([[-8, 1, -3],[1, -2, 1],[k, 7, -4]])

det_A = A.det()

Sol = sp.solve(det_A, k)

print(f'Los valores de k son: {Sol}')

```

```python
# Solución Ejercicio 1. (No olvide imprimir las respuestas de sus cálculos)

# (b):
import sympy as sp

k = sp.Symbol('k')

A = sp.Matrix([[-k, k-1, k+1, k], [1, 2, 3, 4],[2-k, k+3, k+7, 2],[-2, 3, 4, 6]])

det_A = A.det()

Sol = sp.solve(det_A + 7*k -1, k)

print(f'Los valores de k son: {Sol}')
```

### **Ejercicio #4: (1.2)**
Considere el espacio vectorial

$$ D_2 := \left\lbrace \begin{equation*}
 A =
\begin{pmatrix}
 a_{11} & 0 \\
0 & a_{22}
\end{pmatrix}
\end{equation*} :  a_{11},a_{22} \in \mathbb{R} \right\rbrace,$$
es decir, el espacio de las matrices diagonales $2 \times 2$ con las operaciones usuales de las matrices.
Si $A$ y $B$ están en $D_2$, defina el producto escalar y la norma como sigue:

$$  \langle A,B \rangle = a_{11}b_{11} + a_{22}b_{22}, \quad \quad \|A\|  = \sqrt{a_{11}^2 + a_{22}^2}.          $$


  a). **(0.6)** Dadas las matrices \begin{equation*}
  A =
  \begin{pmatrix}
  10 &  0\\
  0 & -1
  \end{pmatrix}, \
  B =
  \begin{pmatrix}
  4 & 0\\
  0 & 6
  \end{pmatrix},
  \end{equation*} halle $\langle A,B \rangle + 2 \langle A^2,-B \rangle$.  



b). **(0.6)** Dadas las matrices \begin{equation*}
  A =
  \begin{pmatrix}
  -59 &  0\\
  0 & -26
  \end{pmatrix}, \
  B =
  \begin{pmatrix}
  275 & 0\\
  0 & 45
  \end{pmatrix}, \
  C =
  \begin{pmatrix}
  9 & 0\\
  0 & -7
  \end{pmatrix},
  \end{equation*} halle $ 4 \langle A + B, C \rangle -2 \| C-B \|$.



```python
#########---------- Ejecutar esta parte antes de dar respuestas a las preguntas-----------#########

# NOTA: El producto escalar aqui definido puede calcularse  mediante la siguiente función:

Product = lambda A,B: np.trace(A @ B)
Norm = lambda A: np.sqrt(np.trace(A @ A))

# Dicha función recibe como entradas las matrices A,B y devuelve como resultado el producto escalar de A y B.

# Ejemplo: Hallemos el producto escalar <A,B> con las matrices definidas a continuación

A = np.array([[9, 0],[0,2]])

B = np.array([[1, 0],[0,3]])

Sol = Product(A,B)

print('Sol =', Sol)
```

### **Ejercicio #5: (1.2)**
Considere el conjunto

$$ \mathbb{R}^{2} := \{ \vec{u} = (u_1,u_2):  u_{1},u_{2} \in \mathbb{R} \}.$$

Para  $\vec{u} = (u_1,u_2)$, $\vec{v} = (v_1,v_2)$ en  $\mathbb{R}^{2}$ y $\alpha \in \mathbb{R}$ se definen  las operaciones:

$$\vec{u} \oplus \vec{v} := (u_1 + v_1 + 2,u_2 + v_2 + 2),$$

$$ \alpha \odot \vec{u} := (\alpha + \alpha u_1 -1,\alpha + \alpha u_2 -1).   $$

a). **(0.6)** Halle $  4 \odot (1,-3) \oplus 9 \odot (-5,2). $

b). **(0.6)** Muestre que el vector $\vec{0} = (-2,-2)$ es el elemento neutro aditivo verificando la igualdad $\vec{0} \oplus (-10,5) = (-10,5).$


```python
#########---------- Ejecutar esta parte antes de dar respuestas a las preguntas-----------#########

# NOTA: Las operaciones suma y producto aqui definidas pueden ser calculadas  mediante las siguientes funciones:

SUMA = lambda u,v:  np.array([[u[0,0] + v[0,0] + 2,u[0,1] + v[0,1] + 2]])

# La función SUMA recibe como entradas los vectores u,v y devuelve como resultado la suma de dichos vectores.

PRODUCTO = lambda alpha,u:  np.array([[alpha + alpha*u[0,0] - 1,alpha + alpha*u[0,1] - 1]])

# La función PRODUCTO recibe como entradas el escalar alpha y el vector u, y devuelve como resultado el producto escalar de alpha por u.

# Ejemplo: Hallemos la suma de u = (2,7) con v = (0,4).

u = np.array([[2, 7]])

v = np.array([[0, 4]])

Sol = SUMA(u,v)

print('Sol =', Sol)
```

```python
# Solución Ejercicio 5. (No olvide imprimir las respuestas de sus cálculos)

# Debe usar en sus respuestas las funciones antes definida

# (a):

u = np.array([[1, -3]])
alpha = 4
result_4u = PRODUCTO(alpha, u)

v = np.array([[-5, 2]])
alpha = 9
result_9v = PRODUCTO(alpha, v)

result_a = SUMA(result_4u, result_9v)
print('Resultado parte a):', result_a)
```
