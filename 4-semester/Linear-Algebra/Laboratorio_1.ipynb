### **Ejercicio #1: (1.0)**
Dados los siguientes vectores

\begin{equation*}
 \vec{u} = (50, -200, 0),
 \hspace{1cm}
 \vec{v} = (8, 609, 12),
\hspace{1cm}
 \vec{w} = (-4, -5, 2),
 \hspace{1cm}
 \vec{z} = (10, 24, 78),
\end{equation*}

realizar las operaciones indicadas:

a). **(0.25)**  $ \vec{u}\cdot \vec{z} + \vec{w}\cdot (3\vec{v}).$

b). **(0.25)**  $ (\vec{u} + \vec{z}- \vec{v}) \cdot (\vec{z}-3\vec{w} + \vec{v}).$

c). **(0.25)**  $\| \vec{z}- \vec{w}\|- \|\vec{w} \times \vec{u}\|.$

d). **(0.25)**  $\|\vec{u}\|(\vec{v} + \vec{z}) + \| 2\vec{z}- 3\vec{w}\|\vec{z}.$




```python
# Solución Ejercicio 1.

# Librerias necesarias para trabajar
# Proceso que se realiza unicamente una vez
import numpy as np
import math

# Declaramos los vectores
u = np.array([50, -200, 0])
v = np.array([8, 609, 12])
w = np.array([-4, -5, 2])
z = np.array([10, 24, 78])

# En todas las operaciones se sigue un patron donde se divide en 2 la operacion
# para un mejor procesamiento y analisis de la operacion.

# (a):

resultado1 = u @ z
resultado2 = w @ (3*v)
respuesta1 = resultado1 + resultado2

print(respuesta1)

# (b):

resultado1 = u+z-v
resultado2 = z - (3*w) + v
respuesta2 = resultado1 @ resultado2

print(respuesta2)

# (c):

resultado1 = np.linalg.norm(z - w)
resultado2 = np.linalg.norm(np.cross(w, u))
respuesta3 = resultado1 - resultado2

print(respuesta3)

# (d):

resultado1 = np.linalg.norm(u) * (v + z)
resultado2 = np.linalg.norm((2*z) - (3*w)) * z
respuesta4 = resultado1 + resultado2

print(respuesta4)

```

### **Ejercicio #2: (0.8)**
 Sean $\vec{u} = (7, 10, -2)$ y $\vec{v} = (2, −4, 1)$. Hallar dos vectores $\vec{p}$ y $\vec{h}$ tales que $\vec{v} = \vec{p} + \vec{h}$ con $\vec{p}$ paralelo a $\vec{u}$ y $\vec{h}$ ortogonal a $\vec{u}$.

```python
# Solución Ejercicio 2.

# Declaramos los vectores
u = np.array([7, 10, -2])
v = np.array([2, -4, 1])

# Calculo la norma de u
norm_u = np.linalg.norm(u)

# Elevo al cuadrado la norma de u
norm_u2 = math.pow(norm_u, 2)

# Hallo el producto punto de u con v
prod_uv = u @ v

# Hallo la constante lambda
Lambda = prod_uv / norm_u2

# Hallo la proyección ortogonal p
p = Lambda * u

# Defino el vector h = v-p
h = v - p

print(p, h)

```

### **Ejercicio #5: (1.5)**
Se tiene una placa rectangular cuyas orillas se mantienen a cierta temperatura, como se puede apreciar en la figura de abajo. Nos interesa encontrar la temperatura en los puntos interiores ($T_1 $ a $T_6$). Suponga que la temperatura en un punto interior es el promedio de la temperatura de los cuatro puntos que lo rodean: arriba, a la derecha, abajo y a la izquierda. Por ejemplo, en el punto $T_1$ se debe satisfacer que $T_1 = \frac{20 + T_4 + 0 + T_2}{4}.$    

\

Temperatura2.png

\

**(a).** **(1.0)** Establezca un sistema de ecuaciones y resuélvalo para econtrar el valor de la temperatura en cada punto interior.

**(b).** **(0.25)** ¿En cuál punto interior la temperatura es máxima?

**(c).** **(0.25)** ¿En cuál punto interior la temperatura es mínima?

```python
# Solución Ejercicio 5.

# (a):

# Coeficientes de las variables
A = np.array([[4, -1, 0, -1, 0, 0], [-1, 4, -1, 0, -1, 0], [0, -1, 4, 0, 0, -1], [-1, 0, 0, 4, -1, 0], [0, -1, 0, -1, 4, -1], [0, 0, -1, 0, -1, 4]])
# Lado derecho de las ecuaciones
B = np.array([20, 20, 35, 0, 0, 15])

# Resuelve el sistema de ecuaciones
solucion = np.linalg.solve(A, B)

print(solucion)

# (b)

# Encuentra el valor máximo del array
valor_maximo = np.max(solucion)

# Imprime el valor máximo
print("Valor máximo:", valor_maximo)

# (c)

# Encuentra el valor mínimo del array
valor_minimo = np.min(solucion)

# Imprime el valor mínimo
print("Valor mínimo:", valor_minimo)
```
